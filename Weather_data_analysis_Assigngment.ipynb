```python
data = pd.read_csv('/content/1. Weather Data.csv')
```

```python
#The first 10 rows
data.head(10)
```

```python
#Finding all records where the weather was exactly clear.
clear_weather = data[data['Weather'] == 'Clear']
print(clear_weather)
```

```python
#Finding the number of times the wind speed was exactly 4km/hr
wind_speed = data[data['Wind Speed_km/h'] == 4]
count_wind_speed_4 = wind_speed.shape[0]
print(f"The wind speed was exactly 4 km/hr {count_wind_speed_4} times.")
```

```python
#To check if there are Null values present in the dataset
data.isnull().sum()
```

```python
#To rename rhe column "weather" to "weather_condition"
data.rename(columns = {'Weather' : 'Weather_condition'}, inplace = True)
data.head()
```

```python
#What is the mean visibility of the dataset?
mean_visibility = data['Visibility_km'].mean()
print(f"The mean visibility of the dataset is {mean_visibility} km.")
```

```python
#Find the number of records where the wind speed is greater than 24km/hr and visibility is equal to 25km
# Filter the records where wind speed is greater than 24 km/hr and visibility is equal to 25 km
filtered_records = data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)]

# Count the number of occurrences
count_filtered_records = filtered_records.shape[0]

# Display the count
print(f"The number of records with wind speed greater than 24 km/hr and visibility equal to 25 km is {count_filtered_records}.")
```

```python
#what is the mean value of each column for each weather condition
mean_values = data.groupby('Weather_condition').mean()
# Display the mean values for each weather condition
print(mean_values)
```

```python
# prompt: what is the mean value of each column for each weather condition

#what is the mean value of each column for each weather condition
mean_values = data.groupby('Weather_condition').mean()
# Display the mean values for each weather condition
print(mean_values)
```

```python
#find all instances where the weather is clear and the relative  humidity is greater than 50, or visibility is above 40
filtered_data = data[(data['Weather_condition'] == 'Clear') & (data['Rel Hum_%'] > 50) | (data['Visibility_km'] > 40)]
print(filtered_data)
```

```python
#find the number of weather conditions that include snow
snow_weather = data[data['Weather_condition'].str.contains('Snow')]
count_snow_weather = snow_weather.shape[0]
print(f"The number of weather conditions that include snow is {count_snow_weather}.")
```
